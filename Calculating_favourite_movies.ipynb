```python
import pandas as pd
comoda = pd.read_csv('LDOS-CoMoDa.csv')
```

```python
print(comoda)
ratings = comoda['rating'].tolist()
print(sum(ratings)/len(ratings))
```

```python
comoda = comoda[['itemID','rating','age','sex']]
print(comoda)
itemIDs = pd.unique(comoda['itemID']).tolist()
print(itemIDs)
```

```python
def getHighestRated(dataframe):
  newList=[]
  itemIDs = pd.unique(dataframe['itemID']).tolist()
  for id in itemIDs:
    temp = dataframe[dataframe['itemID']==id]
    ratings = temp['rating'].tolist()
    temp2 = []
    for r in ratings:
      if r>3:
        temp2.append(r)
    ratings=temp2
    newList.append([id,len(ratings)])
  sorted_list = sorted(newList, key=lambda t: t[1],reverse=True)
  print(sorted_list)
```

```python
getHighestRated(comoda[comoda['age']<18])
```

```python
import statistics
ages = comoda['age'].tolist()
ages = list(filter(lambda a: a != -1, ages))
print(ages)
ages.remove(900)
statistics.median(ages)
print(min(ages))
print(max(ages))
```
